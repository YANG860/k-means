```python
import numpy as np
from matplotlib import pyplot as plt
import random

size = 100
clusters = 10
epoch = 200


x = [random.random() for i in range(size)]
y = [random.random() for i in range(size)]

center_x = [random.random() for i in range(clusters)]
center_y = [random.random() for i in range(clusters)]


def get_distance(x1,y1,x2,y2):    
    return (x1-x2)**2+(y1-y2)**2

for i in range(epoch):
    cluster_index = []
    cluster_size = [0]*clusters
    
    for dot in zip(x,y):
        distance = [get_distance(*dot,*center) for center in zip(center_x,center_y)]                
        index = distance.index(min(distance) )
        cluster_index.append(index)
        cluster_size[index] += 1
        
        
    center_x=[0.0]*clusters
    center_y=[0.0]*clusters
    for j,*dot in zip(cluster_index,x,y):
        center_x[j] += dot[0] /cluster_size[j]
        center_y[j] += dot[1] /cluster_size[j]
        


g = lambda x: x*50 % 255       #用组序号生成一些颜色的参数
color = [g(x) for x in cluster_index]
```

```python

plt.scatter(x,y,c=color),
plt.scatter(center_x,center_y,c="red",marker="*",s=100,)
```
